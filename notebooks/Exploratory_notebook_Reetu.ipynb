```python
final_df_Control['date_time'] = pd.to_datetime(final_df_Control['date_time'], errors='coerce')

# 1. Calculate Total Visit Time (Ensuring next_time conversion if needed)
final_df_Control['next_time'] = pd.to_datetime(final_df_Control.groupby('visit_id')['date_time'].shift(-1), errors='coerce')

final_df_Control['time_diff'] = final_df_Control['next_time'] - final_df_Control['date_time']

# Remove rows with NaT (Not a Time) values in time_diff
final_df_Control.dropna(subset=['time_diff'], inplace=True)

total_visit_time_control = final_df_Control.groupby('visit_id')['time_diff'].sum().reset_index()
total_visit_time_control.rename(columns={'time_diff': 'total_visit_time'}, inplace=True)


# 2. Calculate Time Spent per Step (Adjusted for final_df_Control)

def calculate_time_per_step_control(group):
    start_time_series = group.loc[group['process_step'] == 'start', 'date_time']

    # Check if there's a start time
    if not start_time_series.empty:
        start_time = start_time_series.iloc[0]
        times = {}
        for step in ['step_1', 'step_2', 'step_3', 'confirm']:
            step_times = group.loc[group['process_step'] == step, 'date_time']
            if not step_times.empty:
                end_time = step_times.iloc[-1]  
                times[f'time_{step}'] = (end_time - start_time).total_seconds()
            else:
                times[f'time_{step}'] = pd.Timedelta(0).total_seconds()
        return pd.Series(times)
    else:
        # Handle the case where 'start' is not present 
        return pd.Series({f'time_{step}': pd.Timedelta(0).total_seconds() for step in ['step_1', 'step_2', 'step_3', 'confirm']}) 


time_per_step_control = final_df_Control.groupby('visit_id').apply(calculate_time_per_step_control)


# 3. Merge Results and Add Confirm Check (Adjusted for final_df_Control)
final_output = final_df_Control[['visit_id']].drop_duplicates().merge(
    total_visit_time_control, on='visit_id', how='left'
)  # Use left join to keep all visit_ids

final_output = final_output.merge(
    time_per_step_control, on='visit_id', how='left'
)  # Use left join to keep all visit_ids

final_output['has_confirm'] = final_df_Control.groupby('visit_id')[
    'process_step'
].apply(lambda x: 'confirm' in x).reset_index()[0]

# Fill missing values with 0 (for visits that didn't have all steps)
final_output.fillna(0, inplace=True)

# Select and order columns
final_output = final_output[
    [
        'visit_id',
        'total_visit_time',
        'time_step_1',
        'time_step_2',
        'time_step_3',
        'time_start',
        'time_confirm',
        'has_confirm',
    ]
]
```
